```python
import yfinance as yf
import pandas as pd
import numpy as np
import scipy.stats as si
from nselib import capital_market
```

```python
# Input inversement money value
portfolio_size = float(input('Enter the amount you want to invest: '))

#Fetechh the list of nifty 50 index tickers
data = capital_market.nifty50_equity_list()
nifty_50_tickers = [ticker + '.NS' for ticker in data['Symbol']]

individual_investment = portfolio_size / len(nifty_50_tickers)
my_columns = ['Ticker', 'Current Price','Market capital', 'Stocks to buy (Equal Weight)', 'Stocks to buy (Market Cap Weight)', 'Market_cap_weight%']
data_rows = []
total_market_capital = 0


print(nifty_50_tickers)
```

```python
for ticker in nifty_50_tickers:
    try:
        data = yf.Ticker(ticker)
        data_info = data.info

        current_price = data_info.get('currentPrice', np.nan)
        market_capital = data_info.get('marketCap', np.nan)

        if not np.isnan(market_capital):
            total_market_capital += market_capital

        if not np.isnan(current_price) and not np.isnan(market_capital):
            stocks_can_buy_equal = individual_investment / current_price
            data_rows.append({
                'Ticker': ticker,
                'Current Price': current_price,
                'Market capital': market_capital,
                'Stocks to buy (Equal Weight)': stocks_can_buy_equal,
                'Stocks to buy (Market Cap Weight)': np.nan,
                'Market_cap_weight%': np.nan
            })

    except Exception as e:
        print(f"Could not fetch data for {ticker}: {e}")

for i, row in enumerate(data_rows):
    market_capital = row['Market capital']
    if not np.isnan(market_capital) and total_market_capital > 0:
        market_cap_weight = (market_capital / total_market_capital) * 100
        stocks_can_buy_market_cap = (market_capital / total_market_capital) * portfolio_size / row['Current Price']

        data_rows[i]['Market_cap_weight%'] = market_cap_weight
        data_rows[i]['Stocks to buy (Market Cap Weight)'] = stocks_can_buy_market_cap

data_frame = pd.DataFrame(data_rows, columns=my_columns)
data_frame.head()
```

```python
#Saving file as csv or excel
data_frame.to_csv('nifty_50_data.csv', index=False)
data_frame.to_excel('nifty_50_data.xlsx', index=False)
```
